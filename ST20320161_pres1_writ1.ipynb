**clean data(Missing values+ Imputation**)

This code enabled me to clean the combined allcities dataframe with regard to missing values. It also prevents needless data by not using the columns that are completely empty, and making sure that all of the columns contain the data with the help of imputation of the numerical and categorical columns. Median is a resilient measure against outliers and mode is resilient against skewed data because it would maintain the overall distribution of the dataset.

Inference

The code makes sure that no gaps in the dataset. Columns containing 100 percent nulls are dropped, numerical columns are filled in with the median, and the categorical column AQIBucket is filled in with its mode. This gets allcities prepared to be modeled or analysed and maintain the integrity of the data and avoid making mistakes in calculations.

**Exploratory Data Analysis(EDA)**

```python
import numpy as np
# Drop columns with 100% missing values in all_cities
cols_to_drop_100_percent_missing_all_cities = [col for col in all_cities.columns if all_cities[col].isnull().all()]
if cols_to_drop_100_percent_missing_all_cities:
    print(f"Dropping 100% missing columns in all_cities: {cols_to_drop_100_percent_missing_all_cities}")
    all_cities = all_cities.drop(columns=cols_to_drop_100_percent_missing_all_cities)

# Impute numerical columns with median in all_cities
numerical_cols_all_cities = all_cities.select_dtypes(include=np.number).columns
for col in numerical_cols_all_cities:
    if all_cities[col].isnull().any():
        median_val = all_cities[col].median()
        all_cities[col] = all_cities[col].fillna(median_val)
        print(f"  - Imputed numerical column {col} in all_cities with median: {median_val}")

# Impute categorical column (AQI_Bucket) with mode in all_cities
if 'AQI_Bucket' in all_cities.columns and all_cities['AQI_Bucket'].isnull().any():
    mode_val = all_cities['AQI_Bucket'].mode()[0]
    all_cities['AQI_Bucket'] = all_cities['AQI_Bucket'].fillna(mode_val)
    print(f"  - Imputed categorical column AQI_Bucket in all_cities with mode: {mode_val}")

# Verify no more missing values in all_cities
if all_cities.isnull().sum().sum() == 0:
    print(f"\nNo missing values remaining in all_cities after cleaning.")
else:
    print(f"\nMissing values still present in all_cities after cleaning:")
    print(all_cities.isnull().sum()[all_cities.isnull().sum() > 0])
```

Reason for Output

The lists of output showed dropped columns and reported the imputation used. It ensures that the absent values have been addressed. The following results are achieved due to the following code checks: missing data verified with the help of the .isnull, imputation and details printing, which makes the cleaning process traceable and correct.

How It Produced the Output

The code initializes completely empty columns and removes them. In the case of numerical columns, it determines the median and fills in the gaps; and in categorical columns, the mode is used. Subsequently, after processing, allcities.isnull.sum makes sure that there is no left-over data. print statements give a straight forward record of all cleaning that has been done.

```python
all_cities_data = []

city_files = glob.glob("*_data.csv") # Finding all the files that end with _data.csv
for file_name in city_files:
    city_df = pd.read_csv(file_name) # Reading the CSV file into a data frame
    # Adding city's data to a list
    all_cities_data.append(city_df)

    print(f"Loaded: {file_name}")

# Combining all city data into one big table
combined_data = pd.concat(all_cities_data, ignore_index=True)

# Saving the combined data to a new CSV file
combined_data.to_csv("all_cities_combined.csv", index=False)

print(f"SUCCESS: Combined {len(city_files)} city files into one file with {len(combined_data)} total rows")
print("The combined file is saved as: all_cities_combined.csv")
```


I applied this code to plot average AQI in various air quality buckets. plotly.express gives an opportunity to create interactive and visually beautiful bar charts. The division into AQIBuckets and the mean AQI allows one to view changes in trends based on a category, whereas the establishment of a particular order makes one able to interpret the idea of a good or a severe AQI meaningfully.

Inference

The bar chart indicates mean average AQI of each air quality. This assists in determining the buckets that cause the greatest amount of pollution and also ease comparison. As an illustration, the bigger the bar the lower the quality of air. It is better to use colors corresponding to the level of severity, and then one can get a clear picture of the distribution of the air quality in the cities.

```python
import plotly.express as px
avg_aqi_per_bucket = all_cities.groupby('AQI_Bucket')['AQI'].mean().reset_index()

# Define a specific order for AQI_Bucket for better visualization
aqi_bucket_order = ['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe']
avg_aqi_per_bucket['AQI_Bucket'] = pd.Categorical(avg_aqi_per_bucket['AQI_Bucket'], categories=aqi_bucket_order, ordered=True)
avg_aqi_per_bucket = avg_aqi_per_bucket.sort_values('AQI_Bucket')

fig = px.bar(avg_aqi_per_bucket,
             x='AQI_Bucket',
             y='AQI',
             color='AQI_Bucket',
             title='Average AQI per Air Quality Bucket',
             labels={'AQI': 'Average Air Quality Index', 'AQI_Bucket': 'Air Quality Bucket'},
             category_orders={'AQI_Bucket': aqi_bucket_order},
             color_discrete_map={
                 'Good': 'green',
                 'Satisfactory': 'lime',
                 'Moderate': 'yellow',
                 'Poor': 'orange',
                 'Very Poor': 'darkorange',
                 'Severe': 'red'
             },
             hover_data={'AQI_Bucket': True, 'AQI': ':.2f'})

fig.update_layout(xaxis_title='Air Quality Bucket', yaxis_title='Average AQI')
fig.show()
```

Reason for Output

The result is an interactive bar chart of mean AQI per bucket. AQI results are aggregated with groupby, and the plot is created with px.bar. Categorical order and custom colors generate logical sequence and easy interpretation. Hover data uses accurate numeric values to every bucket, which improves the process of exploratory analysis and whether to communicate the results or not.

How It Produced the Output

allcities.groupby(AQI_Bucket) mean(AQI) of all categories of AQI. The sort and categorical type ensures that the plotting is ordered using the use of color mapping, labels, titles, and hover information that are present on the bar chart created by the use of the figure (fig.bar), and the chart is interactively displayed on the screen, allowing zooming, hovering, and exploration to have an intuitive understanding of the trend of air quality across buckets.

AQI Bucket Distribution (Doughnut Chart)

I operated based off of this code to visualize the distribution of records in AQI buckets. valuecounts() rapidly provides a summary of the distribution of categories, and plotly.express.pie gives an interactive and intuitive display. Categorical ordering and color mapping make the chart readable and easy to analyze as it makes the severity of air quality clear and consistent.

Inference

The pie (donut) chart indicates the percentage of records within every category of air quality. This assists in determining the most common air quality situations among cities. That is, the bigger the slices are, the more frequent they are in terms of AQI categories. The visualization helps to identify the trends in the urban air quality and help to comprehend the distribution of the pollution levels.

```python
aqi_bucket_counts = all_cities['AQI_Bucket'].value_counts().reset_index()
aqi_bucket_counts.columns = ['AQI_Bucket', 'Count']

# Define a specific order for AQI_Bucket for better visualization and consistent coloring
aqi_bucket_order = ['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe']
aqi_bucket_counts['AQI_Bucket'] = pd.Categorical(aqi_bucket_counts['AQI_Bucket'], categories=aqi_bucket_order, ordered=True)
aqi_bucket_counts = aqi_bucket_counts.sort_values('AQI_Bucket')

fig = px.pie(aqi_bucket_counts,
             values='Count',
             names='AQI_Bucket',
             title='Distribution of Air Quality Buckets',
             hole=0.4, # Creates the donut chart effect
             color='AQI_Bucket', # Color by AQI_Bucket
             color_discrete_map={
                 'Good': 'green',
                 'Satisfactory': 'lime',
                 'Moderate': 'yellow',
                 'Poor': 'orange',
                 'Very Poor': 'darkorange',
                 'Severe': 'red'
             },
             hover_data=['AQI_Bucket', 'Count'], # Corrected from dictionary to list
             labels={'AQI_Bucket': 'Air Quality Bucket', 'Count': 'Number of Records'})

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Time Series Plot of AQI by Date

I was able to use this code to track the AQI patterns over time in cities. To sort and plot properly, Date to datetime conversion will be used then plotly.express.line() will generate an interactive and multi-city line chart so that it is easier to compare time trends and see peak and plateau periods of pollution in each city at once, effectively.

Inference

The line chart indicates time series AQI of each city. It emphasizes the seasonal or temporary patterns and makes it possible to compare the cities. Peaks show the times of air pollution and troughs denote cleaner air. This visualization can be used to identify trends, identify abnormalities and to see the way pollution changes over time across cities.

```python
# Convert 'Date' column to datetime objects
all_cities['Date'] = pd.to_datetime(all_cities['Date'], format='%d/%m/%Y', errors='coerce')

# Drop rows where 'Date' conversion failed (if any)
all_cities.dropna(subset=['Date'], inplace=True)

# Sort data by Date for proper time series plotting
all_cities.sort_values(by='Date', inplace=True)

# Create the line plot for AQI over time, colored by City
fig = px.line(
    all_cities,
    x='Date',
    y='AQI',
    color='City',
    title='Time Series Trend of AQI by City',
    labels={'AQI': 'Air Quality Index', 'Date': 'Date'},
    hover_data={'City': True, 'Date': True, 'AQI': True}
)

fig.update_layout(
    xaxis_title="Date",
    yaxis_title="Air Quality Index (AQI)",
    hovermode="x unified"
)

fig.show()
```

Reason for Output

The result is an interactive graph of a line chart, with AQI on the y-axis and Date on the x-axis. The by date sorting would make the lines chronological. It is very easy to interpret and compare trends over time because coloring by city and hover data enables users to see the values at the specific date.

How It Produced the Output

It is possible to change strings into datetime objects using pdtodatetime and then sort properly using sortvalues, then plot AQI vs Date with color per city using pxlindemas, and finally polish the layout using updatelayout; therefore, we can now meaningfully plot time.

Train-Test Split

To prepare the dataset to be used in machine learning, I applied this code to split features (X) and target (y) into the training and testing sets, and traintestsplit guarantees that the training and testing of the machine learning model remain unbiased (stratify=y).

Inference

The code produces model training and testing datasets. 80% of the data serves to train the model and 20% to test the model. This makes sure that the model is learning enough examples and has enough unseen data to test the performance. Stratification ensures that there is proportional representation of AQI in both sets.


```python
from sklearn.model_selection import train_test_split

# Drop non-numeric and irrelevant columns
X = all_cities[['PM2.5','PM10','NO','NO2','NOx','NH3',
        'CO','SO2','O3','Benzene','Toluene','Xylene','AQI']]


y = all_cities['AQI_Bucket_encoded']  # Target variable


X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% test data
    random_state=42,    # reproducible split
    stratify=y          # ensures proportional target distribution
)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
print("Training target shape:", y_train.shape)
print("Test target shape:", y_test.shape)
```

Reason for Output

The output presents the sample and features in the training and test sets. The shapes verify that the split has been done correctly. stratify=y makes the proportions of the target classes the same in both sets to avoid skewed training or evaluation scores, and a representative sample of the dataset is available to the model is being learned.

How It Produced the Output

X and AQIBucketencoded are numeric columns of a selected set of numeric pollutants and the target is the y, traintestsplit randomly divides the data and preserves the target distribution using stratify=y. The split is legitimized by printing shapes. The resultant Xtrain, Xtest, train y, and y test are available in pipelines of model training and evaluation.

KNN Model,Random Forest Model,XGBoost Model

I referred to K-Nearest Neighbors (KNN) since it is a simple instance based classifier, which predicts the target according to the majority class of the nearest neighbors. It is effective in multi-class issues such as AQI buckets and simple to apply when initial baseline modelling is required.

Inference

The code returns accuracy and a report on the classification of the predictions of KNN on the test set. The accuracy reflects general correct predictions, whereas the classification report presents precision, recall, and F1-score of each bucket of AQI. This can be used to evaluate the model performance in the various categories of air quality, as well as the strengths and weaknesses.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

#  KNN CLASSIFIER

# Impute NaNs in X_train and X_test with their medians before training KNN
X_train_imputed = X_train.fillna(X_train.median())
X_test_imputed = X_test.fillna(X_test.median())

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_imputed, y_train)

y_pred_knn = knn.predict(X_test_imputed)

print(" KNN Classifier")
print("Accuracy: ", accuracy_score(y_test, y_pred_knn))
print("\n Classification Report:\n", classification_report(y_test, y_pred_knn))
```

Reason for Output

The output indicates the performance of the KNN model on the unseen data to classify the buckets of the AQI, with accuracyscore giving the overall performance, and a classificationreport giving the performance by each of the classes. The difference in metrics also shows the buckets that can and can not be predicted easily which in many cases depends on the distribution of classes or the separation of features.

How It Produced the Output

The default KNeighborsClassifier(nneighbors=5) is initialized with 5 neighbors, fit uses it with Xtrain and ytrain, and predict making predictions with Xtest, and accuracyscore and classificationreport provide the quantitative measures of performance of the model to use and compare with other models.

# KNN


I applied this code to visualize the KNN model performance by AQI buckets. Transforming the classification report into a dataframe and visualizing the precision, recall and F1-score values per bucket can help determine which air quality category the model predicts well or not. This makes it a multi-class evaluation that is more interpretable.

Inference

The grouped bar chart presents the precision, recall and F1-score of each bucket of AQI. High scores are a sign of high prediction performance, whereas low scores indicate buckets of misclassifications.
```python
import pandas as pd
import plotly.express as px
from sklearn.metrics import classification_report

# Generate classification report for KNN
report_knn = classification_report(y_test, y_pred_knn, output_dict=True)

# Convert report to DataFrame for plotting, excluding 'accuracy', 'macro avg', 'weighted avg'
report_df_knn = pd.DataFrame(report_knn).transpose()
report_df_knn = report_df_knn.drop(columns=['support'])
report_df_knn = report_df_knn.drop(['accuracy', 'macro avg', 'weighted avg'], errors='ignore')

# Add an 'AQI_Bucket' column for proper labeling
# Map encoded labels back to original AQI_Bucket names for better readability
report_df_knn['AQI_Bucket_encoded'] = report_df_knn.index.astype(int)

# Create a mapping from encoded value to original label
label_mapping = {code: bucket for bucket, code in zip(le_bucket.classes_, le_bucket.transform(le_bucket.classes_))}
report_df_knn['AQI_Bucket'] = report_df_knn['AQI_Bucket_encoded'].map(label_mapping)

# Melt the DataFrame for grouped bar chart
report_melted_knn = report_df_knn.melt(id_vars=['AQI_Bucket', 'AQI_Bucket_encoded'], var_name='Metric', value_name='Score')

# Define a specific order for AQI_Bucket for better visualization
aqi_bucket_order = ['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe']
report_melted_knn['AQI_Bucket'] = pd.Categorical(report_melted_knn['AQI_Bucket'], categories=aqi_bucket_order, ordered=True)
report_melted_knn = report_melted_knn.sort_values('AQI_Bucket')

# Create the grouped bar chart
fig = px.bar(report_melted_knn,
             x='AQI_Bucket',
             y='Score',
             color='Metric',
             barmode='group',
             title='KNN Model Performance: Precision, Recall, and F1-Score per AQI Bucket',
             labels={'AQI_Bucket': 'Air Quality Bucket', 'Score': 'Score'},
             hover_data={'Metric': True, 'Score': ':.2f'})

fig.update_layout(
    xaxis_title='Air Quality Bucket',
    yaxis_title='Score',
    yaxis_range=[0, 1.05] # Set y-axis range to 0-1 for scores
)

fig.show()
```

Model Evaluation(Accuracy & Precision Comparison Plots)

This code was used by me when comparing the overall performance of KNN, Random Forest, and XGBoost models. Accuracy and weighted precision are complementary measures: the accuracy demonstrates overall accuracy whereas weighted precision considers imbalances in the classes. These metrics can be easily compared using Plotly to compare across models in a short period of time.

Inference

The bar charts indicate variations in model performance. Greater accuracy and precision denote more effective general predictions. Weighted precision makes sure that the buckets of AQI whose occurrence is less frequent are taken into account. This aids in determining the model that is most effective in general and determines the model to be deployed or refined in terms of predicting the AQI categories.

```python
from sklearn.metrics import accuracy_score, classification_report, precision_score

# Calculate Accuracy & Precision for each model
accuracy_scores = {
    'KNN': accuracy_score(y_test, y_pred_knn),
    'Random Forest': accuracy_score(y_test, y_pred_rf),
    'XGBoost': accuracy_score(y_test, y_pred_xgb)
}

precision_scores = {
    'KNN': precision_score(y_test, y_pred_knn, average='weighted', zero_division=0),
    'Random Forest': precision_score(y_test, y_pred_rf, average='weighted', zero_division=0),
    'XGBoost': precision_score(y_test, y_pred_xgb, average='weighted', zero_division=0)
}

# Convert to DataFrame for Plotly
acc_df = pd.DataFrame(list(accuracy_scores.items()), columns=['Model', 'Accuracy'])
prec_df = pd.DataFrame(list(precision_scores.items()), columns=['Model', 'Precision'])

# Plot Accuracy using Plotly
fig_acc = px.bar(acc_df,
                 x='Model',
                 y='Accuracy',
                 color='Model', # Color each bar by model
                 title='Model Accuracy Comparison',
                 labels={'Accuracy': 'Accuracy Score'},
                 hover_data={'Model': True, 'Accuracy': ':.4f'})
fig_acc.update_layout(yaxis_range=[0, 1])
fig_acc.show()

# Plot Precision using Plotly
fig_prec = px.bar(prec_df,
                  x='Model',
                  y='Precision',
                  color='Model', # Color each bar by model
                  title='Model Precision Comparison',
                  labels={'Precision': 'Precision Score (Weighted)'},
                  hover_data={'Model': True, 'Precision': ':.4f'})
fig_prec.update_layout(yaxis_range=[0, 1])
fig_prec.show()
```


I applied this code to construct an interactive Streamlit dashboard to predict EDA and AQI buckets. Streamlit is a Python data app which can be deployed quickly. Plotly offers interactive displays. The ability to load saved models and processed data is necessary to ensure consistency between training and prediction of the predictions, which makes the app user-friendly and reproducible.

Inference

The application features two key features, including an EDA dashboard to visualize air quality trends and distributions and a model prediction interface that allows the user to enter values of pollutants to forecast AQI buckets. The users are able to intuitively plot city-scaled AQI, pollutant distributions, and time dynamics in addition to testing the ML models in real-time.

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import joblib

# ---------------------------------------------
# PAGE SETUP
# ---------------------------------------------
st.set_page_config(
    page_title="AQI Dashboard",
    page_icon="",
    layout="wide"
)

# ---------------------------------------------
# LOAD DATA & MODELS
# ---------------------------------------------
@st.cache_data
def load_data():
    df = pd.read_csv("all_cities.csv")
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    return df

@st.cache_resource
def load_models():
    knn = joblib.load("knn_model.pkl")
    rf = joblib.load("rf_model.pkl")
    xgb = joblib.load("xgb_model.pkl")
    encoder = joblib.load("label_encoder.pkl")
    return knn, rf, xgb, encoder

df = load_data()
knn, rf, xgb, encoder = load_models()

# ---------------------------------------------
# SIDEBAR NAVIGATION
# ---------------------------------------------
st.sidebar.title("Navigation")
page = st.sidebar.radio("Go to", [" EDA Dashboard", " Model Prediction"])


# =========================================================
#              PAGE 1: EDA DASHBOARD
# =========================================================
if page == " EDA Dashboard":

    st.title(" Exploratory Data Analysis (EDA)")
    st.markdown("Interactive visual analytics for Air Quality Dataset")

    st.subheader("Dataset Preview")
    st.write(df.head())

    # ----------------- City Count -----------------
    st.subheader("City Data Count")
    city_counts = df["City"].value_counts()
    st.bar_chart(city_counts)

    # ----------------- AQI Bucket Pie Chart -----------------
    st.subheader("AQI Bucket Distribution")
    bucket_counts = df["AQI_Bucket"].value_counts()

    fig = px.pie(
        names=bucket_counts.index,
        values=bucket_counts.values,
        title="AQI Category Distribution",
        hole=0.4
    )
    st.plotly_chart(fig, use_container_width=True)

    # ----------------- AQI Time Series -----------------
    st.subheader("AQI Time Series Trend")
    fig = px.line(
        df,
        x="Date",
        y="AQI",
        color="City",
        title="AQI Trend Over Time"
    )
    st.plotly_chart(fig, use_container_width=True)

    # ----------------- Pollutant Boxplot -----------------
    pollutants = ['PM2.5', 'PM10', 'NO2', 'O3']
    chosen_pollutant = st.selectbox("Select Pollutant for Boxplot", pollutants)

    fig = px.box(
        df,
        x="City",
        y=chosen_pollutant,
        color="City",
        title=f"{chosen_pollutant} Distribution Across Cities"
    )
    st.plotly_chart(fig, use_container_width=True)



# =========================================================
#              PAGE 2: MODEL PREDICTION
# =========================================================
else:

    st.title(" AQI Bucket Prediction System")

    st.markdown("""
    Enter pollutant levels to predict the **AQI Category**
    Models available: **KNN, Random Forest, XGBoost**
    """)

    # ---- EXACT FEATURES USED DURING TRAINING ----
    feature_names = [
        'PM2.5','PM10','NO','NO2','NOx','NH3',
        'CO','SO2','O3','Benzene','Toluene','Xylene','AQI'
    ]

    inputs = {}

    st.subheader("Enter Pollutant Values")
    cols = st.columns(3)

    # 3-column input layout
    for i, feat in enumerate(feature_names):
        with cols[i % 3]:
            inputs[feat] = st.number_input(feat, min_value=0.0, step=0.1)

    # Select Model
    model_choice = st.radio(
        "Choose a Model",
        ["KNN", "Random Forest", "XGBoost"],
        horizontal=True
    )

    # Predict Button
    if st.button("Predict AQI Bucket"):

        values = np.array(list(inputs.values())).reshape(1, -1)

        if model_choice == "KNN":
            pred = knn.predict(values)
        elif model_choice == "Random Forest":
            pred = rf.predict(values)
        else:
            pred = xgb.predict(values)

        result = encoder.inverse_transform(pred)[0]

        st.success(f"###  Predicted AQI Bucket: **{result}**")
        st.info("Prediction completed successfully!")
```

Reason for Output

The product is an interactive web site. The dataset can be explored using visualizations such as bar charts, pie charts, line plots and box plots. A prediction page takes pollutant levels that are introduced by the user and predicts the AQI bucket through pre-trained models. The combination offers predictive functionality as well as analytical insights.

How It Produced the Output

st.pageconfig allows to configure the page. st.cachedata and st.cache resource are efficient to load data and models. There is page navigation through the sidebar. Plotly generates EDA charts on features that are selected by a user. The prediction page is used to input the numeric values, choose a model, predict AQI bucket and map the encoded predictions to their original labels through the usage of LabelEncoder.
